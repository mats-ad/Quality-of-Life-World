```python
import pandas as pd # type: ignore
```

```python
LifeExpectancy = pd.read_csv('Data/LifeExpectancy.csv')
LifeExpectancy
```

```python
import plotly.express as px

# Weltkarte mit Plotly erstellen
fig = px.choropleth(
    LifeExpectancy,
    locations="Country",           # Spalte mit Ländern
    locationmode="country names",  # Ländercodes oder Namen
    color="Life expectancy",                 # Spalte mit Werten
    title="Weltkarte mit Werten pro Land",
    color_continuous_scale="Viridis"  # Farbskala
)

# Werte in die Länder schreiben (als Text-Overlay)
'''for i, row in LifeExpectancy.iterrows():
    fig.add_scattergeo(
        locations=[row["Country"]],
        locationmode="country names",
        text=f"{row['Life expectancy']}",     # Wert anzeigen
        mode="text",
        showlegend=False
    )
'''
# Karte anzeigen
fig.show()
```
