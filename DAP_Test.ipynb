```python
CostOfLiving = pd.read_csv('Data/Cost_of_Living_Index_by_Country_2024.csv')
CostOfLiving
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Daten basierend auf 'Land' zusammenführen
merged_data = pd.merge(CostOfLiving, LifeExpectancy, on='Country')

# Nur die Werte für das Clustering extrahieren
values = merged_data[['Cost of Living Index', 'Life expectancy']]

# Daten normalisieren (Standardisierung)
scaler = StandardScaler()
scaled_values = scaler.fit_transform(values)

# Finden der optimalen Anzahl an Clustern mit der Elbow-Methode
wcss = []  # Within-Cluster Sum of Squares
silhouette_scores = []
cluster_range = range(2, 11)

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_values)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(scaled_values, kmeans.labels_))

# Elbow-Methode visualisieren
plt.figure(figsize=(10, 5))
plt.plot(cluster_range, wcss, marker='o')
plt.title('Elbow-Methode zur Bestimmung der optimalen Clusteranzahl')
plt.xlabel('Anzahl der Cluster')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()

# Silhouette-Score visualisieren
plt.figure(figsize=(10, 5))
plt.plot(cluster_range, silhouette_scores, marker='o')
plt.title('Silhouette-Score für unterschiedliche Clusteranzahlen')
plt.xlabel('Anzahl der Cluster')
plt.ylabel('Silhouette-Score')
plt.show()

# Optimale Anzahl an Clustern basierend auf Silhouette-Score
optimal_clusters = cluster_range[silhouette_scores.index(max(silhouette_scores))]
print(f'Optimale Anzahl an Clustern: {optimal_clusters}')

# K-Means mit optimaler Clusteranzahl
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
merged_data['Cluster'] = kmeans.fit_predict(scaled_values)

# Ergebnisse anzeigen
print(merged_data)

# Clustering visualisieren
plt.scatter(merged_data['Cost of Living Index'], merged_data['Life expectancy'], c=merged_data['Cluster'], cmap='viridis')
plt.xlabel('Cost of Living Index')
plt.ylabel('Life expectancy')
plt.title('Clustering von Ländern')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import pandas as pd

# Beispiel-Datenrahmen (ersetze diese mit deinen tatsächlichen Tabellen)
table1 = LatestLifeExpectancy
table2 = LatestWordHappiness
table3 = CrimeRate
table4 = UnemploymentRate
table5 = WaterQuality
table6 = GDP
table7 = CostOfLiving

# Liste der Tabellen und jeweilige Spalten mit Länderdaten
tables = [
    (table1, 'Country'),
    (table2, 'Country name'),
    (table3, 'country'),
    (table4, 'name'),
    (table5, 'country'),
    (table6, 'country'),
    (table7, 'Country')
]

# Initialisieren des kombinierten DataFrames
combined_table = tables[0][0].rename(columns={tables[0][1]: 'Country'})

# Iteratives Joinen der Tabellen
for table, country_column in tables[1:]:
    table_renamed = table.rename(columns={country_column: 'Country'})  # Länder-Spalte umbenennen
    combined_table = pd.merge(
        combined_table,
        table_renamed,
        on='Country',
        how='outer',  # Vollständiges äußeres Joinen
        suffixes=('', f'_{country_column}')  # Präfix für Konflikte
    )

# Ergebnis anzeigen
combined_table.reset_index(drop=True, inplace=True)
combined_table
```

```python
combined_table.columns
```

```python
# New column order
new_order = ['Country', 'Year', 'Life expectancy', 'year', 'Life Ladder',
       'crimeRateByCountry_crimeIndex', 'date_of_information', '%',
       'WaterQuality_EPIDrinkingWaterScore_2024', 'gdpPerCapita',
       'Cost of Living Index', 'continent']

# Reorder the columns
combined_table = combined_table[new_order]
```

```python
combined_table = combined_table.rename(columns={'Year': 'Year_Life_expectancy', 'year': 'Year_Life_Ladder',
       'crimeRateByCountry_crimeIndex': 'Crime Rate', 'date_of_information': 'Year_Unemployment_Rate', '%': 'Unemployment_Rate',
       'WaterQuality_EPIDrinkingWaterScore_2024': 'Water_Quality'})
combined_table
```
